```python
import os

import numpy as np
import pandas as pd

from IPython.display import display

DIR = r'c://downloads'
```

```python
viperdb = pd.read_csv(os.path.join(DIR, 'viperdb.csv'))
display(viperdb)
```

```python
display(viperdb.loc[viperdb['Outer Radius'].idxmax()])
```

```python
from scipy.stats import pearsonr

def calc_correlation(array1, array2):
    null_indices = pd.isnull(array1) | pd.isnull(array2)
    return pearsonr(array1[~null_indices], array2[~null_indices])

def print_correlation(array1, array2):
    print('r = %.3f, p = %.1e' % calc_correlation(array1, array2))

print_correlation(viperdb['Inner Radius'], viperdb['Outer Radius'])
```

```python
dsDNA_records = viperdb[viperdb['Genome'].fillna('').str.startswith('dsDNA')]
ssRNA_records = viperdb[viperdb['Genome'].fillna('').str.startswith('ssRNA')]
print('dsDNA:')
display(dsDNA_records.describe())
print('ssRNA:')
display(ssRNA_records.describe())
```

dsDNS viruses are larger than ssRNA according to virtually all measures.

```python
new_cancer_cases_by_state = pd.read_csv(os.path.join(DIR, 'new_cancer_cases_by_state.csv'), thousands = ',', index_col = 'State')
display(new_cancer_cases_by_state)
```

```python
print((new_cancer_cases_by_state['Leukemia'] / new_cancer_cases_by_state['All Sites']).idxmax())
```

```python
normalized_cancer_types = new_cancer_cases_by_state.drop(columns = 'All Sites').div(new_cancer_cases_by_state['All Sites'], \
        axis = 0).drop('United States')
display(normalized_cancer_types)
```

```python
print('Unclassified cancer per state:')
print(1 - normalized_cancer_types.sum(axis = 1))
```

```python
best_r = None
best_match = None

for state1, profile1 in normalized_cancer_types.iterrows():
    for state2, profile2 in normalized_cancer_types.iterrows():
        if state1 < state2:
            
            r, _ = calc_correlation(profile1, profile2)
            
            if best_r is None or r > best_r:
                best_r = r
                best_match = (state1, state2)
                
print('Best correlation is between %s and %s with r = %.5f' % (best_match + (best_r,)))
```
